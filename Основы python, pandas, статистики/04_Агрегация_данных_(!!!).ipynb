```python
prod_cols = {'quantity': 'sum'}

df \
 .groupby('product') \
 .agg(prod_cols) \
 .reset_index()
```

```python
prod_cols = {'quantity': 'sum'}

(df
 .groupby(['product'])
 .agg(prod_cols)
 .reset_index()
)
```

```python
prod_cols = {'quantity':
             ['sum', 'mean', 'std', 'max']}

(df
 .groupby(['product'])
 .agg(prod_cols)
)
```

```python
prod_cols = {'quantity':
             ['sum', 'mean', 'std', 'max', 'count']}

(
df
 .groupby(['company', 'product'])
 .agg(prod_cols)
)
```

```python
df_gr = (df
          .groupby(['company', 'product'])
          .agg(prod_cols)
          .fillna(0)
          )

#df_gr
#df_gr.xs("Abatz", drop_level=False)
#df_gr.xs("Abatz")

#df_gr["quantity"]["sum"].values[:4]
#df_gr["quantity"]["sum"].idxmax()
```

```python
(df
 .groupby(['company', 'product'])
 .agg(prod_cols)
 .reset_index()
)
```

```python
(df
 .groupby(['company'])
 .agg({'invoice': 'count',
      'extended amount': 'max'})
 .rename(columns={"invoice": "invoice_total",
                  'extended amount': "max_purchase"})
)
```

```python
(df
 .groupby(['company'])
 .agg(invoice_total=('invoice', 'count'),
      max_purchase=('extended amount', 'max'))
)
```

```python
(df
 .groupby(['company'])
 .agg({'invoice': 'count',
       'extended amount': 'max'})
)
```

## Pivot table and crosstab

```python
df.head()
```

```python
pd.pivot_table(df,
               index=['company'],
               columns=['product'],
               values=['extended amount'],
               aggfunc='sum')
```

```python
pd.pivot_table(df,
               index=['company'],
               columns=['product'],
               values=['extended amount'],
               aggfunc='sum',
               margins=True, # итог All
               fill_value=0) # заполняет нулями
```

```python
pd.pivot_table(df,
               index=['company'],
               columns=['product'],
               values=['extended amount'],
               aggfunc=['sum', 'mean', 'max'],
               margins=True,
               fill_value=0)
```

```python
pd.pivot_table(df,
               index=['company', 'product'],
               values=['extended amount'],
               aggfunc=['sum'],
               margins=True,
               fill_value=0)
```

```python
pd.crosstab(df['company'],
            df['product']) # считает кол-во элементов по умолчанию
```

```python
pd.crosstab(df['company'],
            df['product'],
            values=df['extended amount'], # что считаем
            aggfunc='sum', # как считаем
            normalize='index') # показывает процент от общих покупок
```

```python
pd.crosstab(df['company'],
            df['product'],
            values=df['extended amount'], # что считаем
            aggfunc='sum', # как считаем
            normalize='columns') # показывает процент от общих покупок
```
