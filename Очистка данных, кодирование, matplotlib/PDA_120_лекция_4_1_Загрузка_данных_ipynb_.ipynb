## Working with data types (***РАБОТА С ТИПАМИ ДАННЫХ при загрузке Excel***)

```python
df_levels.info()
```

```python
df_levels.head(2) # до применения  parse_dates={'update_date': [3,4,5]
```

```python
df_levels = pd.read_excel(src_file,
                          sheet_name='levels',
                          header=2,
                          usecols='A,C:H',
                          parse_dates={'update_date': [3,4,5]})
# parse_dates={'update_date': [3,4,5] - (укзываются индексы) столбики 'день','месяц','год'
# преобразуются в один столбик с типом данных 'datetime64[ns]' (см. ниже)
```

```python
df_levels.head(2) # после применения  parse_dates={'update_date': [3,4,5]
```

```python
df_levels.info()
```

```python
df_levels.head() # непорядок с ZipCode
```

```python
df_levels = pd.read_excel(src_file,
                          sheet_name='levels',
                          header=2,
                          usecols='A,C:H',
                          parse_dates={'update_date': [3,4,5]},
                          dtype={'ZipCode': 'object'})
# делаем ZipCodeб где все значения 5-ти значные
# dtype={'ZipCode': 'object'} при переводе столбца 'ZipCode' в 'object' впереди
# подставляется '0'!!!
```

```python
df_levels.head()
```

```python
df_levels.info()
```

```python
def convert_percent(val):
    new_val = val.replace('%', '')
    return float(new_val)/100
# функция убирает %, преобразует в float и делим на 100
```

```python
convert_percent("87%")
```

```python
df_levels['pct_to_target']
```

```python
df_levels['pct_to_target'].apply(convert_percent).head(2)
# apply -  применяет функцию к столбику !!!!!!!!!!!!!!!!
```

```python
df_levels = pd.read_excel(src_file,
                          sheet_name='levels',
                          header=2,
                          usecols='A,C:H',
                          parse_dates={'update_date': [3,4,5]},
                          dtype={'ZipCode': 'object'},
                          converters={'pct_to_target': convert_percent})
# converters={'pct_to_target': convert_percent} - при загрузке данных
# converters - дает возможность применять функцию к столбцу
# в данном случае применили функцию convert_percent
```
